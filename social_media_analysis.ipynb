# Project 3: Social Media Usage Analysis

## Overview
This analysis focuses on the dataset **percent-of-men-and-women-using-social-media-platforms-in-the-us.csv**.  
The goal is to explore how social media usage can be used as a proxy for social connection and how such information could be misused in an unethical profiling system.

## Why this dataset
This dataset is useful because it shows differences in social media platform usage across groups. In the context of our project, social media usage can be treated as a signal of social connectedness, and lower engagement may suggest weaker support networks or higher vulnerability.

```python
import pandas as pd

df = pd.read_csv("percent-of-men-and-women-using-social-media-platforms-in-the-us.csv")
df.head()
```

```python
men_col = "Social media use among men in the US (%)"
women_col = "Social media use among women in the US (%)"

df_use = df[["Entity", "Year", men_col, women_col]]

df_use = df_use[df_use[men_col].notna() | df_use[women_col].notna()]

df_use.head()
```

```python
df_use.describe()
```

```python
df_use[[
    "Social media use among men in the US (%)",
    "Social media use among women in the US (%)"
]].mean()
```

```python
import matplotlib.pyplot as plt

df_use.set_index("Entity")[[
    "Social media use among men in the US (%)",
    "Social media use among women in the US (%)"
]].plot(kind="bar")

plt.title("Social Media Usage by Gender")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.show()
```

## Visualization Analysis

The chart shows clear differences in social media usage across platforms and between genders. Platforms such as YouTube and Facebook have the highest usage rates for both men and women, indicating that they are widely used across the population.

There are also noticeable gender differences on certain platforms. For example, women tend to have higher usage on platforms like Facebook, Instagram, and Pinterest, while men show relatively higher usage on Reddit. These differences suggest that user behavior varies across platforms and demographic groups.

```python
df_use["difference"] = (
    df_use["Social media use among women in the US (%)"] -
    df_use["Social media use among men in the US (%)"]
)

df_use.set_index("Entity")["difference"].plot(kind="bar")

plt.title("Gender Difference in Social Media Usage (Women - Men)")
plt.ylabel("Percentage Difference")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Additional Insight

The difference between male and female usage highlights which platforms have the largest gender gaps. Platforms with larger differences may allow companies to more easily segment users and apply targeted strategies, increasing the risk of exploitation.

## Vulnerability Signals

Social media usage can be used as a proxy for social connection. Individuals with lower engagement on social media platforms may have weaker social networks and fewer support systems, making them more vulnerable to manipulation or unfair treatment.

Additionally, differences in platform usage across genders suggest that companies could design targeted strategies to exploit specific groups. For example, users on certain platforms may be more susceptible to targeted advertising or influence, allowing organizations to apply different pricing, messaging, or decision-making strategies based on behavioral patterns.

This demonstrates how even simple behavioral data can be used to build profiles of vulnerability in an unethical system.

## Conclusion

This analysis demonstrates how social media usage data can reveal behavioral patterns across platforms and demographic groups. While such data can be useful for understanding user behavior, it can also be misused to identify and exploit vulnerable populations in unethical ways.
